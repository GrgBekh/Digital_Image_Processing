```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.models as models
from torch.utils.data import DataLoader, random_split
from torchvision.datasets import CIFAR10
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
```

```python
import numpy as np
```

```python
torch.manual_seed(42)
```

```python
labels_of_classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

full_train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
```

```python
torch.cuda.is_available()
```

```python
train_size = int(0.8 * len(full_train_dataset))
validation_size = len(full_train_dataset) - train_size
train_dataset, validation_dataset = random_split(full_train_dataset, [train_size, validation_size])

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Model, Loss, and Optimizer
device = 'cuda'
resnet_model = models.resnet18(pretrained=False, num_classes=10).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(resnet_model.parameters(), lr=0.001)
```

```python
num_epochs = 10
model_with_sgd_train_losses = []
model_with_sgd_val_losses = []

for epoch in range(num_epochs):
    # Training phase
    resnet_model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)
    model_with_sgd_train_losses.append(train_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss}")

    # Validation phase
    resnet_model.eval()
    validation_loss = 0.0
    with torch.no_grad():
        for inputs, labels in validation_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet_model(inputs)
            loss = criterion(outputs, labels)
            validation_loss += loss.item()

    validation_loss = validation_loss / len(validation_loader)
    model_with_sgd_val_losses.append(validation_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Validation Loss: {validation_loss}")
```

```python
resnet_model.eval()

true_labels = []
predicted_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = resnet_model(images)
        _, predicted = torch.max(outputs, 1)
        true_labels.extend(labels.cpu().numpy())
        predicted_labels.extend(predicted.cpu().numpy())
```

```python
confusion = confusion_matrix(true_labels, predicted_labels)
total_samples_per_class = confusion.sum(axis=1)
print("accuracy per class: ", confusion.diagonal()/total_samples_per_class)
print("precision per class: ", precision_score(true_labels,  predicted_labels, average = None))
print("recall per class: ", recall_score(true_labels,  predicted_labels, average = None))
print("f1 per class: ", f1_score(true_labels,  predicted_labels, average = None))
```

```python
sns.heatmap(confusion, cbar = False, annot = True)
```

```python
resnet_model = models.resnet18(pretrained=False, num_classes=10).to(device)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(resnet_model.parameters(), lr=0.001)

# Training loop
```

```python
num_epochs = 8
model_with_Adam_train_losses = []
model_with_Adam_val_losses = []

for epoch in range(num_epochs):
    # Training phase
    resnet_model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)
    model_with_Adam_train_losses.append(train_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss}")

    # Validation phase
    resnet_model.eval()
    validation_loss = 0.0
    with torch.no_grad():
        for inputs, labels in validation_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet_model(inputs)
            loss = criterion(outputs, labels)
            validation_loss += loss.item()

    validation_loss = validation_loss / len(validation_loader)
    model_with_Adam_val_losses.append(validation_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Validation Loss: {validation_loss}")
```

```python
resnet_model.eval() 

true_labels = []
predicted_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = resnet_model(images)
        _, predicted = torch.max(outputs, 1)
        true_labels.extend(labels.cpu().numpy())
        predicted_labels.extend(predicted.cpu().numpy())

        
confusion = confusion_matrix(true_labels, predicted_labels)
```

```python
confusion = confusion_matrix(true_labels, predicted_labels)
total_samples_per_class = confusion.sum(axis=1)
print("accuracy per class: ", confusion.diagonal()/total_samples_per_class)
print("precision per class: ", precision_score(true_labels,  predicted_labels, average = None))
print("recall per class: ", recall_score(true_labels,  predicted_labels, average = None))
print("f1 per class: ", f1_score(true_labels,  predicted_labels, average = None))
```
